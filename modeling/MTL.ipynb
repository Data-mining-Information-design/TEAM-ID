```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation
from keras.utils import np_utils
from keras import metrics
```

```python
data = pd.read_csv('/content/drive/My Drive/hynix_short.csv', index_col=0, parse_dates=[0], dayfirst=False,infer_datetime_format=True, encoding='cp949')
```

```python
print(data.isna().sum())
print(data.shape)
print(list(data.columns))
```

```python
data = data.dropna(axis=0)
```

```python
data_x = data[['S종가', 'H종가', '환율', '유가종가', '유가오픈', '유가고가', '유가저가', '유가변동률', '한국기준금리', '미국기준금리', '한미기준금리차이', '반도체수출금액지수', 'SOX종가', 'SOX변동률', 'BTC종가', 'BTC변동률', '코스피종가', '코스피고가', '코스피저가', '코스피거래량', '코스피변동', '코스닥종가', '코스닥시가', '코스닥고가', '코스닥저가', '코스닥거래량', '코스닥변동', 'SP종가', 'SP변동률', '시장금리']]
data_y = pd.concat([data['S등락'], data['H등락']], axis=1)
```

```python
print(data_y.isna().sum())
```

```python
data_y.head()
```

```python
epoch = [32, 64, 128]
batch_sizes = [8, 16, 32]
```

```python
for i in epoch:
    for j in batch_sizes:
        model_file_name = "mt_model_"+str(i)+"_"+str(j)
        print("Model: "+model_file_name)
        print("Epoch: "+str(i)+", Batch size: "+str(j))

        model = Sequential()
        model.add(Dense(128, input_shape=(30,), activation='relu', kernel_initializer='normal'))
        model.add(Dropout(rate=0.1))
        model.add(Dense(128, input_shape=(30,), activation='relu', kernel_initializer='normal'))
        model.add(Dense(128, activation='relu', kernel_initializer='normal'))
        model.add(Dense(data_y.shape[1], activation='sigmoid', kernel_initializer='normal'))

        model.compile(loss='binary_crossentropy', optimizer='adam',  metrics=['accuracy'])

        model.fit(data_x, data_y, epochs=i, batch_size=j, validation_split=0.2, verbose=2)

        y = model.predict(data_x)

        model.save(model_file_name)
        print()
```
